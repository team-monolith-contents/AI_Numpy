### **2. Numpy array와 사칙연산하기**

Numpy의 배열은 파이썬의 리스트보다 강력한 사칙연산 기능을 지원합니다. 먼저 두가지의 경우에 대해 사칙연산이 가능함을 배워보겠습니다.

#### **2.1. 같은 모양을 가지는 배열 간 사칙연산**

먼저, 배열의 모양이 같은 경우 사칙연산은 배열의 각 원소별로 이루어집니다.



위의 그림과 같이 2 x 2 배열 a와 b간 연산을 진행한다면, 어떤 연산을 진행하든 같은 색깔의 칸에 있는 숫자끼리 이루어집니다. 예를 들어, 더하기 연산을 진행한다면 아래와 같은 결과가 나오게 됩니다.




실제 코딩으로도 결과를 확인해볼까요? 우선 더할 경우에 결과를 확인해보아요.
1. 위의 그림과 같이 2x2 배열 a와 b를 만들어주세요.
2. 일반 변수처럼 두 변수를 더해주세요.

```python
# 예제 1: 배열간의 사칙연산 덧셈
a = np.array([[2, 4], [6, 8]])
b = np.array([[1, 2], [3, 4]])
print(a + b)
```

a에서 b를 뺀 경우에도 결과를 확인해보아요.

```python
# 예제 2: 배열간의 사칙연산 뺄셈
print(a - b)
```

a에서 b를 곱한 경우에도 결과를 확인해보아요.

```python
# 예제 3: 배열간의 사칙연산 곱셈
print(a * b)
```

a에서 b를 나눈 경우에도 결과를 확인해보아요.

```python
## 예제 4: 배열간의 사칙연산 나눗셈
print(a / b)
```

### **2.2. Numpy array와 숫자 간 사칙연산**
